```python
def predict(train, test, predictors, model):
    model.fit(train[predictors], train["target"])
    preds = model.predict(test[predictors])
    preds = pd.Series(preds, index=test.index, name="predictions")
    combined = pd.concat([test["target"], preds], axis=1)
    return combined
```

```python
def backtest(data, model, predictors, start=1095, step=150):
    all_predictions = []

    for i in range(start, data.shape[0], step):
        train = data.iloc[0:i].copy()
        test = data.iloc[i:(i+step)].copy()
        predictions = predict(train, test, predictors, model)
        all_predictions.append(predictions)
    
    return pd.concat(all_predictions)
```

```python
from xgboost import XGBClassifier

model = XGBClassifier(random_state=1, learning_rate=.1, n_estimators=1000)
predictions = backtest(eth, model, predictors)
```

```python
def compute_rolling(eth) :
    horizions = [3,7,60,365]
    new_predictors = ["close","sentiment_x","neg_sentiment_x"]
    
    for horizon in horizons :
        rolling_averages = eth.rolling(horizon,min_period=1).mean()
        
        ratio_column = f"close_ratio_{horizon}"
        eth[ratio_column] = eth["close"] / rolling_averages["close"]
        
        edit_column = f"edit_{horizon}"
        eth[edit_column] = rolling_averages["edit_count_x"]
        
        rolling = eth.rolling(horizon,closed="left",min_periods=1).mean()
        
        trend_column = f"trend_{horizon}"
        eth[trend_column] = rolling["target"]
        
    
        new_predictors+= [edit_column,ratio_column,trend_column]
        
    return eth , new_predictors   
```

```python
eth , new_predictors = compute_rolling(eth.copy())
```
